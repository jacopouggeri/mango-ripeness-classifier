```python
import pandas as pd
import numpy as np
import mlxtend
import ast
from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import association_rules
from mlxtend.preprocessing import TransactionEncoder
df = pd.read_csv("filtered_ingred.csv")
```

```python
#df['Filtered_Text']=df.Filtered_Text.str.split(',').values.tolist()
df=df['Filtered_Text'].values
```

```python
initial_df=df.tolist()
result=np.array(initial_df,dtype=object)
```

```python
result
```

```python
for i in range(len(result)):
    result[i]=ast.literal_eval(result[i])
```

```python
result
```

```python
from mlxtend.preprocessing import TransactionEncoder
te = TransactionEncoder()
transaction_matrix = te.fit_transform(result, sparse=True)
transaction_matrix
```

```python
transaction_df = pd.DataFrame.sparse.from_spmatrix(transaction_matrix, columns = te.columns_)
transaction_df.head()
```

```python
mlxtend.__version__
```

```python
#feed into apriori
x = apriori(transaction_df, min_support = 0.02, use_colnames=True,low_memory=True)
```

```python
x.sort_values(by = 'support', ascending = False)
```

```python
rules_df = association_rules(x, metric = 'lift', min_threshold=1)
rules_df.sort_values(by='lift',ascending = False)
```

```python
rules_df['antecedents'] = rules_df['antecedents'].apply(lambda x:list(x)).copy()
rules_df['consequents'] = rules_df['consequents'].apply(lambda x:list(x)).copy()
```

```python
import numpy as np
# Input basket
mybasket = ['salt', 'apple', 'sugar','butter']

#metric
metric = 'lift'

#COMPLETE THIS FUNCTION
def product_recs(basket, rule_df, metric):
    
    # Randomly select an item from the basket
    random_item = np.random.choice(basket, 1)[0]
    print(random_item)
    
    # Find rules where the item is in the antecedent
    rule_filter = rule_df['antecedents'].apply(lambda x:x[0]) == random_item
    
    # Filter the dataframe using rule_filter and sort by the selected metric
    filtered_df = rule_df[rule_filter].sort_values(by = metric, ascending = False)
    
    # Randomly return one of the top 20 items from the filtered dataframe
    reco = filtered_df.head(20).sample()['consequents']
    
    return reco

product_recs(mybasket, rules_df, metric )
```
